```python
'''Creates a third column in the .csv file with the processed tweets'''
import pandas as pd
import spacy
import nltk
from nltk.tokenize import word_tokenize

# Load Spacy's English-language model
nlp = spacy.load("en_core_web_sm")

# Ensure NLTK's tokenizers and Spacy's model are downloaded
nltk.download('punkt')

# Function to tokenize and lemmatize text
def tokenize_and_lemmatize(text):
    # Tokenize the text
    tokens = word_tokenize(text)

    # Convert tokens to Spacy's Document to perform lemmatization
    doc = nlp(" ".join(tokens))

    # Lemmatize each token and compile back into a single string
    lemmatized_text = " ".join([token.lemma_ for token in doc])

    return lemmatized_text

# Apply the tokenize_and_lemmatize function to the tweet column and create a new column with the processed text
df_twitter['processed_tweets'] = df_twitter['clean_text'].apply(tokenize_and_lemmatize)

# Display the first few rows to verify
print(df_twitter.head())

```

```python
import pandas as pd
import nltk

file_path_dict = '/content/drive/My Drive/APS360 Group/Data preprocessing/Text data/dictionary.txt'
file_path_label = '/content/drive/My Drive/APS360 Group/Data preprocessing/Text data/sentiment_labels.txt'


# Load phrases and their IDs, assuming dictionary.txt does not have headers
phrases_df = pd.read_csv(file_path_dict, sep='|', names=['phrase', 'phrase_id'])
phrases_df.set_index('phrase_id', inplace=True)

# Load phrase ids and their corresponding sentiment scores
sentiment_scores_df = pd.read_csv(file_path_label, sep='|')
sentiment_scores_df.rename(columns={'phrase ids': 'phrase_id', 'sentiment values': 'sentiment_score'}, inplace=True)
sentiment_scores_df.set_index('phrase_id', inplace=True)

# Merge the DataFrames
rt_df = phrases_df.join(sentiment_scores_df)

rt_df
```

```python
# Convert sentiment scores to labels
def score_to_label(score):
    if score < 0.4:
        return -1
    elif score <= 0.6:
        return 0
    else:
        return 1

rt_df['sentiment_label'] = rt_df['sentiment_score'].apply(score_to_label)

# Reset index to make 'phrase_id' a column if needed for further processing
# Note: If you do not need 'phrase_id' in your final DataFrame, you can skip the reset_index step.
final_df = rt_df.reset_index()[['phrase', 'sentiment_label']]

# Filter rows where 'Data' contains at least one alphabetic character
rt_df = rt_df[rt_df['phrase'].str.contains('[a-zA-Z]', regex=True)]

nltk.download('words')
from nltk.corpus import words
english_words_set = set(words.words())

# Function to check if a phrase contains at least one English word
def contains_english_word(s):
    tokens = s.split()
    return any(token.lower() in english_words_set for token in tokens)

# Further filter to remove rows that do not contain valid English words
rt_df = rt_df[rt_df['phrase'].apply(contains_english_word)] #.set_index('ID')

rt_df = rt_df.drop(['sentiment_score'], axis =1)

rt_df
```

```python
# Augmenting the data

import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from tensorflow.keras.utils import save_img

# Define the path to the source folder
source_folder = '/content/drive/My Drive/APS360 Group/Data preprocessing/Image data 2/Highly positive'

# Define the path for the augmented images to be saved
# Ensure this directory exists or create it with os.makedirs()
augmented_folder = '/content/drive/My Drive/APS360 Group/Data preprocessing/Image data 2/Augmented_Positive'

if not os.path.exists(augmented_folder):
    os.makedirs(augmented_folder)

# Initialize the ImageDataGenerator with various transformations
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Loop through each image in the source folder
for filename in os.listdir(source_folder):
    if filename.endswith(".png") or filename.endswith(".jpg") or filename.endswith(".jpeg"):
        image_path = os.path.join(source_folder, filename)
        img = load_img(image_path)  # Load image
        x = img_to_array(img)  # Convert to numpy array
        x = x.reshape((1,) + x.shape)  # Reshape to (1, width, height, channels)

        # Generate and save 5 augmented samples for each image
        i = 0
        for batch in datagen.flow(x, batch_size=1,
                                  save_to_dir=augmented_folder,
                                  save_prefix='aug',
                                  save_format='jpeg'):
            i += 1
            if i >= 5:
                break  # Enough augmentations

print("Data augmentation completed.")
```

```python
'''prepares the images for the data augmentation'''
from PIL import Image
import numpy as np

# Define the preprocessing function
def preprocess_images(folder_path, size=(224, 224)):
    for subdir, dirs, files in os.walk(folder_path):
        for file in files:
            try:
                img_path = os.path.join(subdir, file)
                img = Image.open(img_path).convert('RGB')
                img = img.resize(size, Image.ANTIALIAS)

                # Convert to numpy array for pixel scaling
                img_array = np.array(img)
                # Centering pixel values (mean subtraction)
                img_array = img_array - np.mean(img_array, axis=(0,1), keepdims=True)
                # Optional: normalization (divide by std deviation)
                img_array = img_array / np.std(img_array, axis=(0,1), keepdims=True)

                # Save the preprocessed image
                img = Image.fromarray(np.uint8(img_array))
                img.save(img_path)
            except Exception as e:
                print(f"Error processing {file}: {e}")

final_folders = {
    'Positive_final': 5000,
    'Negative_final': 5000,
    'Neutral_final': 5000
}

# Preprocess images in final folders
for final_folder in final_folders.keys():
  preprocess_images(os.path.join(base_path, final_folder'))
```

```python
from torchvision import transforms, datasets

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((244, 244)),
    transforms.Normalize(mean=[0.4393, 0.4045, 0.3695], std=[0.3005, 0.2908, 0.2983])
])

# Create an ImageFolder instance aka a dataloader
dataset = datasets.ImageFolder(root='/content/drive/My Drive/APS360 Group/Data preprocessing/Image data 2/Reduced data', transform=transform)
```

```python
import numpy as np
from torch.utils.data.sampler import SubsetRandomSampler
import torch


def get_loader(dataset, batch_size = 64):
  # Shuffle and split the training set from the rest
    np.random.seed(1000)
    indices = list(range(len(dataset)))  # Generate indices for the entire dataset
    np.random.shuffle(indices)
    split_train = int(len(indices) * 0.8)  # 80-20 split
    train_indices, test_val_indices = indices[:split_train], indices[split_train:]

    # Shuffle and split the rest into testing and validation sets
    np.random.shuffle(test_val_indices)
    split_test_val = int(len(test_val_indices) * 0.5)  # 50-50 split
    test_indices, val_indices = test_val_indices[:split_test_val], test_val_indices[split_test_val:]

    train_sampler = SubsetRandomSampler(train_indices)
    val_sampler = SubsetRandomSampler(val_indices)
    test_sampler = SubsetRandomSampler(test_indices)

    train_loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, sampler=train_sampler, num_workers=1)
    print("len train_loader ", len(train_loader))

    val_loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, sampler=val_sampler, num_workers=1)
    print("len val_loader ", len(val_loader))

    test_loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, sampler=test_sampler, num_workers=1)
    print("len test_loader ", len(test_loader))

    return train_loader, val_loader, test_loader

train_loader, val_loader, test_loader, = get_loader(dataset, batch_size = 64)
```

```python
# Feature Extraction with a Pre-trained Model

import torch
import torchvision.models as models
import torch.nn as nn

# Load a pre-trained ResNet18 model
model = models.resnet18(pretrained=True)

# Remove the final fully connected layer to get features instead of predictions
model = nn.Sequential(*list(model.children())[:-1])

# Set the model to evaluation mode
model.eval()

# Function to extract features
def extract_features(loader):
    features = []
    labels = []

    with torch.no_grad():
        for inputs, labels_batch in loader:
            # Assuming your DataLoader returns images and labels
            outputs = model(inputs)
            features.extend(outputs.cpu().numpy().reshape(outputs.size(0), -1))
            labels.extend(labels_batch.cpu().numpy())


    return np.array(features), np.array(labels)

# Extract features for each set
train_features, train_labels = extract_features(train_loader)
val_features, val_labels = extract_features(val_loader)
test_features, test_labels = extract_features(test_loader)
```

```python
# Dimensionality Reduction with PCA
from sklearn.decomposition import PCA

# Instantiate PCA, choose the number of components such as 50 for a start
pca = PCA(n_components=50)

# Fit on training features
train_features_reduced = pca.fit_transform(train_features)

# Transform validation and test features
val_features_reduced = pca.transform(val_features)
test_features_reduced = pca.transform(test_features)
```

```python
# Clustering with K-means
from sklearn.cluster import KMeans

# Choose the number of clusters, K, possibly based on prior knowledge or exploration
k = 3  # For example, aiming to find a rough correspondence to the three sentiment classes

# Perform K-means clustering
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(train_features_reduced)

# Get cluster labels for datasets
train_cluster_labels = kmeans.predict(train_features_reduced)
val_cluster_labels = kmeans.predict(val_features_reduced)
test_cluster_labels = kmeans.predict(test_features_reduced)
```
