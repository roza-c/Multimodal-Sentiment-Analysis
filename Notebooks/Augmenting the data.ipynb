```python
'''Augmenting negative data'''

# Augmenting the data
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from tensorflow.keras.utils import save_img

# Define the path to the Negative folder
source_folder = '/content/drive/My Drive/APS360 Group/Data preprocessing/No Augmentation Images/Final folders/Negative_final' #CHANGE PATH

# Define the path for the augmented images to be saved
augmented_folder = '/content/drive/My Drive/APS360 Group/Data preprocessing/No Augmentation Images/Augmented folders/Augmented_Negative' #CHANGE PATH

if not os.path.exists(augmented_folder):
    os.makedirs(augmented_folder)

datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Loop through each image in the source folder
for filename in os.listdir(source_folder):
    if filename.endswith(".png") or filename.endswith(".jpg") or filename.endswith(".jpeg"):
        image_path = os.path.join(source_folder, filename)
        img = load_img(image_path)  # Load image
        x = img_to_array(img)  # Convert to numpy array
        x = x.reshape((1,) + x.shape)  # Reshape to (1, width, height, channels)

        # Generate and save 5 augmented samples for each image
        i = 0
        for batch in datagen.flow(x, batch_size=1,
                                  save_to_dir=augmented_folder,
                                  save_prefix='aug',
                                  save_format='jpeg'):
            i += 1
            if i >= 5:
                break  # Enough augmentations

print("Data augmentation completed.")
```

```python
'''Move the augmented data points to the original folders that each sentiment class has 10,000 data points'''

import os
import shutil

# Base path where the folders are located
base_path = '/content/drive/My Drive/APS360 Group/Data preprocessing/No Augmentation Images/' #CHANGE PATH

# Source folders
positive_folder_source = os.path.join(base_path, 'Final folders/Positive_final')
negative_folder_source = os.path.join(base_path, 'Final folders/Negative_final')
neutral_folder_source = os.path.join(base_path, 'Final folders/Neutral_final')

# Destination folders
positive_folder_destination = os.path.join(base_path, 'Augmented folders/Positive')
negative_folder_destination = os.path.join(base_path, 'Augmented folders/Negative')
neutral_folder_destination = os.path.join(base_path, 'Augmented folders/Neutral')

augmented_positive = os.path.join(base_path, 'Augmented folders/Augmented_Positive')
augmented_negative = os.path.join(base_path, 'Augmented folders/Augmented_Negative')
augmented_neutral = os.path.join(base_path, 'Augmented folders/Augmented_Neutral')

# Function to move files from source to destination until destination contains a certain number of files
def move_files_to_reach_count(source_folder, dest_folder, target_count):
    existing_files = os.listdir(dest_folder)
    num_files_to_move = target_count - len(existing_files)

    if num_files_to_move <= 0:
        print(f"{dest_folder} already has {target_count} or more files.")
        return

    for file in os.listdir(source_folder)[:num_files_to_move]:
        src_file_path = os.path.join(source_folder, file)
        dest_file_path = os.path.join(dest_folder, file)
        shutil.move(src_file_path, dest_file_path)

# move positive images
for item in os.listdir(positive_folder_source):
  shutil.copy(os.path.join(positive_folder_source, item), positive_folder_destination)

if len(os.listdir(positive_folder_destination)) < 2000:
    move_files_to_reach_count(augmented_positive, positive_folder_destination, 2000)

# Move negative images
for item in os.listdir(negative_folder_source):
  shutil.copy(os.path.join(negative_folder_source, item), negative_folder_destination)

if len(os.listdir(negative_folder_destination)) < 2000:
    move_files_to_reach_count(augmented_negative, negative_folder_destination, 2000)

# Move neutral images
for item in os.listdir(neutral_folder_source):
  shutil.copy(os.path.join(neutral_folder_source, item), neutral_folder_destination)

if len(os.listdir(neutral_folder_destination)) < 2000:
    move_files_to_reach_count(augmented_neutral, neutral_folder_destination, 2000)

print("Folders have been populated according to the specified requirements.")
```
